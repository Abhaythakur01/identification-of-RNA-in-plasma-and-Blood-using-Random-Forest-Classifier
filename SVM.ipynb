```python
import os
import glob

# Specify the directory where you uploaded your CSV files
data_directory = "uploaded_folder"  # Update with the name of your uploaded folder
output_directory = "processed_data"  # Choose a name for your output directory

def remove_zero_value_points(input_file_path, output_file_path):
    with open(input_file_path, 'r') as infile, open(output_file_path, 'w') as outfile:
        for line in infile:
            values = line.strip().split(',')
            if len(values) == 2 and float(values[1]) != 0:  # Check the second column for zero values
                outfile.write(line)

# Create output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Search for CSV files in the data directory and its subdirectories
csv_files = glob.glob(os.path.join(data_directory, '**/*.csv'), recursive=True)

# Process each CSV file
for file_path in csv_files:
    file_name = os.path.basename(file_path)
    print("Processing file:", file_name)  # Print the name of the file being processed

    # Process the file and save processed data to output directory
    output_file_path = os.path.join(output_directory, file_name)
    remove_zero_value_points(file_path, output_file_path)
    print("Processed data saved to:", output_file_path)  # Print the path where processed data is saved

print("Data processing complete. Processed files saved in the output directory.")
```
