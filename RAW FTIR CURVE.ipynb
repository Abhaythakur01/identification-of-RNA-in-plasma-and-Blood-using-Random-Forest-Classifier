```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv("labeled_RNA 100 ug Blood Processed.CSV")

# Extract the relevant columns
wavenumber = data.iloc[:, 0]  # Assuming wave number is in the first column
Absorbance = data.iloc[:, 1]  # Assuming Absorbance is in the second column

# Plot transmission vs. wavenumber
plt.figure(figsize=(10, 6))
plt.plot(wavenumber, Absorbance, color='red')
plt.title('Absorbance vs.Wavenumber')
plt.xlabel('Wavenumber')
plt.ylabel('Absorbance')
plt.grid(True)
plt.show()
```
