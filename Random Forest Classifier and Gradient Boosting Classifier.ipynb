```python
import pandas as pd
import os


data_directory = "uploaded_folder" 
output_directory = "processed_data"  

def remove_zero_value_points(data):
    # Removing rows where Absorbance value is zero
    data = data[data.iloc[:, 1] != 0]
    return data

# Creating output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Iterating through uploaded files in the data directory
for file_name in os.listdir(data_directory):
    file_path = os.path.join(data_directory, file_name)
    if os.path.isfile(file_path) and file_name.endswith('.csv'):
        # Load the data without considering column names
        data = pd.read_csv(file_path, header=None)
        
        # Call function to remove zero-value data points
        data = remove_zero_value_points(data)
        
        # Save processed data to output directory
        output_file_path = os.path.join(output_directory, file_name)
        data.to_csv(output_file_path, index=False, header=False)

print("Data processing complete. Processed files saved in the output directory.")

```

```python
import os
import glob

# Specify the directory where you uploaded your CSV files
data_directory = "uploaded_folder"  
output_directory = "processed_data"  

def remove_zero_value_points(input_file_path, output_file_path):
    with open(input_file_path, 'r') as infile, open(output_file_path, 'w') as outfile:
        for line in infile:
            values = line.strip().split(',')
            if len(values) == 2 and float(values[1]) != 0:  # Check the second column for zero values
                outfile.write(line)

# Create output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Search for CSV files in the data directory and its subdirectories
csv_files = glob.glob(os.path.join(data_directory, '**/*.csv'), recursive=True)

# Process each CSV file
for file_path in csv_files:
    file_name = os.path.basename(file_path)
    print("Processing file:", file_name)  # Print the name of the file being processed
    
    # Process the file and save processed data to output directory
    output_file_path = os.path.join(output_directory, file_name)
    remove_zero_value_points(file_path, output_file_path)
    print("Processed data saved to:", output_file_path)  # Print the path where processed data is saved

print("Data processing complete. Processed files saved in the output directory.")
```

```python
import pandas as pd
import os

# Function to assign compound labels based on wavenumber ranges
def assign_compound_label(wavenumber):
 #  compound labels and wavenumber ranges
    compound_labels = {
        (2850, 3000): 1.0,  # Alkanes
        (1900, 2000): 2.0,  # Alkenes
        (880, 995): 2.0,    # Alkenes
        (3300, 3300): 3.0,  # Alkynes
        (2100, 2250): 3.0,  # Alkynes
        (690, 900): 4.0,    # Arenes
        (3200, 3550): 5.0,  # Alcohols and Phenols
        (1550, 1650): 6.0,  # Amines
        (1720, 1740): 7.0,  # Aldehydes and Ketones
        (1710, 1720): 7.0,  # Aldehydes and Ketones
        (1690, 1690): 7.0,  # Aldehydes and Ketones
        (1675, 1675): 7.0,  # Aldehydes and Ketones
        (1745, 1745): 7.0,  # Aldehydes and Ketones
        (1780, 1780): 7.0,  # Aldehydes and Ketones
        (2500, 3300): 8.0,  # Carboxylic acid and derivatives
        (1705, 1720): 8.0,  # Carboxylic acid and derivatives
        (1210, 1320): 8.0,  # Carboxylic acid and derivatives
        (1785, 1815): 8.0,  # Carboxylic acid and derivatives
        (1750, 1820): 8.0,  # Carboxylic acid and derivatives
        (1040, 1100): 8.0,  # Carboxylic acid and derivatives
        (1735, 1750): 8.0,  # Carboxylic acid and derivatives
        (1000, 1300): 8.0,  # Carboxylic acid and derivatives
        (1630, 1695): 8.0,  # Carboxylic acid and derivatives
        (2240, 2280): 9.0,  # Nitriles
        (2100, 2270): 10.0  # Isocyanates, isothiocyanates, diimides, azides, ketenes
        # Add other compound labels and wavenumber ranges here
    }
    
    
    # Iterate through compound labels and check wavenumber range
    for wavenumber_range, compound_label in compound_labels.items():
        if wavenumber >= wavenumber_range[0] and wavenumber <= wavenumber_range[1]:
            return compound_label
    return 0.0  # Label for wavenumber outside of defined ranges

# Specify the directory containing the CSV files
input_directory_path = "processed_data"  

# Create a new directory for labeled data if it doesn't exist
output_directory_path = "labeled_data"
if not os.path.exists(output_directory_path):
    os.makedirs(output_directory_path)

# List of file names
file_names = [
    "Blood Processed.CSV",
    "Blood Raw.CSV",
    "Plasma + 10ng RNA Processed.CSV",
    "Plasma + 10ng RNA raw.CSV",
    "Plasma + 1mg RNA Processed.CSV",
    "Plasma + 1mg RNA raw.CSV",
    "Plasma Processed.CSV",
    "Plasma Raw.CSV",
    "RNA 1 mg + Blood Processed.CSV",
    "RNA 1 mg + Blood Raw.CSV",
    "RNA 1 mg-drying.CSV",
    "RNA 1 mg.CSV",
    "RNA 1 ug Blood Processed.CSV",
    "RNA 1 ug Blood Raw.CSV",
    "RNA 1 ug Dry Processed (12 hrs).CSV",
    "RNA 1 ug Dry Raw (12 hrs).CSV",
    "RNA 1 ug Plasma Processed.CSV",
    "RNA 1 ug Plasma Raw.CSV",
    "RNA 10 ng + Blood Processed.CSV",
    "RNA 10 ng Dry Processed.CSV",
    "RNA 10 ng Dry Raw.CSV",
    "RNA 10 ug Blood Processed.CSV",
    "RNA 10 ug Blood Raw.CSV",
    "RNA 10 ug Dry Processed.CSV",
    "RNA 10 ug Dry Raw.CSV",
    "RNA 10 ug Plasma Processed.CSV",
    "RNA 10 ug Plasma Raw.CSV",
    "RNA 100 ng Blood Processed.CSV",
    "RNA 100 ng Blood Raw.CSV",
    "RNA 100 ng Dry Processed.CSV",
    "RNA 100 ng Dry Raw.CSV",
    "RNA 100 ng Plasma Processed.CSV",
    "RNA 100 ng Plasma Raw.CSV",
    "RNA 100 ug Blood Processed.CSV",
    "RNA 100ng Wet.CSV",
    "RNA 100ug Blood Raw.CSV",
    "RNA 100ug Dry Processed.CSV",
    "RNA 100ug Dry Raw.CSV",
    "RNA 100ug Plasma Raw.CSV",
    "RNA 100ug Wet.CSV",
    "RNA 10ug Wet.CSV",
    "RNA 1mg Dry Processed.CSV",
    "RNA 1mg Dry Raw.CSV",
    "RNA 1ug Wet.CSV"
]

# Process and label data files
for file_name in file_names:
    file_path = os.path.join(input_directory_path, file_name)
    try:
        # Load the data without headers
        data = pd.read_csv(file_path, header=None)
        
        # Filter out rows with zero Absorbance values (assuming the second column is Absorbance)
        data_filtered = data[data.iloc[:, 1] != 0]
        
        # Apply the labeling function to assign compound labels based on wavenumber
        data_filtered['Compound_Label'] = data_filtered.iloc[:, 0].apply(assign_compound_label)
        
        # Save the updated data with compound labels to a new CSV file in the labeled_data directory
        output_file_path = os.path.join(output_directory_path, "labeled_" + file_name)
        data_filtered.to_csv(output_file_path, index=False)
        
        print("Data processed and labeled. Output saved to:", output_file_path)
    except Exception as e:
        print("Error processing file:", file_path)
        print(e)
```
